```python
from scipy.optimize import newton 

def f(x):
    # Corrige el término x**3 - 3 * x**2 + x - 1
    return x**3 - 3 * x**2 + x - 1

def fprime(x): 
    # Derivada de la función f(x)
    return 3 * x**2 - 6 * x + 1 

# Llamada al método de Newton
root = newton(f, x0=10, fprime=fprime)

print(root)
```

```python
from scipy.optimize import newton
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**3 - 3 * x**2 + x - 1

def fprime(x):
    return 3 * x**2 - 6 * x + 1

# Encontrar la raíz usando el método de Newton
root = newton(f, x0=10, fprime=fprime)

# Crear un rango de valores de x para graficar la función
x_vals = np.linspace(-3, 4, 400)
y_vals = f(x_vals)

# Graficar la función f(x)
plt.plot(x_vals, y_vals, label='f(x) = $x^3 - 3x^2 + x - 1$', color='blue')

# Marcar la raíz en el gráfico
plt.plot(root, f(root), 'ro', label=f'Raíz: x = {root:.4f}')

# Añadir etiquetas y título
plt.axhline(0, color='black',linewidth=1)  # Línea horizontal (eje x)
plt.axvline(0, color='black',linewidth=1)  # Línea vertical (eje y)
plt.title('Gráfico de la función f(x) con la raíz marcada')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()

# Mostrar el gráfico
plt.grid(True)
plt.show()
```
