```python
#### a) construya el polinomio por minimos cuadrados del grado 1 y calcule el error 

import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Datos
x_vals = [4.0, 4.2, 4.5, 4.7, 5.1, 5.5, 5.9, 6.3, 6.8, 7.1]
y_vals = [102.56, 130.11, 113.18, 142.05, 167.53, 195.14, 224.87, 256.73, 299.50, 326.72]

# Definición de la función lineal
def linear_func(x, slope, intercept):
    return slope * x + intercept

# Ajuste de parámetros
params, _ = curve_fit(linear_func, x_vals, y_vals)
slope, intercept = params

# Creación de la función ajustada
def fitted_func(x):
    return slope * x + intercept

# Calcular puntos de la línea ajustada
x_interp = np.linspace(3, 8, 100)
y_interp = fitted_func(x_interp)

# Cálculo del error cuadrático
def calculate_squared_error(observed, predicted_func, x_data):
    total_error = 0
    for i in range(len(observed)):
        total_error += (observed[i] - predicted_func(x_data[i])) ** 2
    return total_error

# Visualización
plt.scatter(x_vals, y_vals, color='blue', label="Datos")
plt.plot(x_interp, y_interp, color='red', label="Línea ajustada")
plt.title("Ajuste Lineal")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.legend()
plt.show()  

# Resultados
print(f'{round(slope, 3)} x + {round(intercept, 3)}')
print(f'Error: {calculate_squared_error(y_vals, fitted_func, x_vals)}')
```

```python
## 1b construya el polinomio por minimos cuadrados del grado 2 y calcule el error 
# Definir la función cuadrática
def quadratic_eq(x, b2, b1, b0):
    return b2 * x**2 + b1 * x + b0

# Ajuste del modelo cuadrático
coeffs, _ = curve_fit(quadratic_eq, x_vals, y_vals)
b2, b1, b0 = coeffs

# Crear la función ajustada
def fitted_quadratic(x):
    return b2 * x**2 + b1 * x + b0

# Generar datos interpolados
x_fit = np.linspace(3, 8, 100)
y_fit = fitted_quadratic(x_fit)


# Graficar los datos y la curva ajustada
plt.scatter(x_vals, y_vals, color='blue', label="Datos")
plt.plot(x_fit, y_fit, color='red', label="Curva ajustada")
plt.title("Ajuste Cuadrático")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.grid(True)
plt.legend()
plt.show()

# Imprimir resultados
print(f'{round(b2, 3)} x^2 + {round(b1, 3)} x + {round(b0, 3)}')
print(f'Error: {calculate_squared_error(y_vals, fitted_quadratic, x_vals)}')
```

```python
### 1c construya el polinomio por minimos cuadrados del grado 3 y calcule el error 
# Definición de la función cúbica
def cubic_function(x, coef3, coef2, coef1, coef0):
    return coef3 * x**3 + coef2 * x**2 + coef1 * x + coef0

# Ajuste del modelo
coefficients, _ = curve_fit(cubic_function, x_vals, y_vals)
coef3, coef2, coef1, coef0 = coefficients

# Definir la función ajustada
def adjusted_cubic(x):
    return coef3 * x**3 + coef2 * x**2 + coef1 * x + coef0

# Generar valores interpolados
x_interp = np.linspace(3, 8, 100)
y_interp = adjusted_cubic(x_interp)

# Graficar los datos y la curva ajustada
plt.scatter(x_vals, y_vals, color="blue", label="Datos")
plt.plot(x_interp, y_interp, color="red", label="Curva ajustada")
plt.title("Ajuste Cúbico")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.grid(True)
plt.legend()
plt.show()

# Imprimir resultados
print(f'{round(coef3, 3)} x^3 + {round(coef2, 3)} x^2 + {round(coef1, 3)} x + {round(coef0, 3)}')
print(f'Error: {calculate_squared_error(y_vals, adjusted_cubic, x_vals)}')
```

```python
#### 1d construya el polinomio por minimos cuadrados del grado 4 y calcule el error 
# Definir la función exponencial
def exponential_func(x, c1, c2):
    return c1 * np.exp(c2 * x)

# Ajustar el modelo exponencial
coefficients, _ = curve_fit(exponential_func, x_vals, y_vals)
c1, c2 = coefficients

# Crear la función ajustada
def fitted_exponential(x):
    return c1 * np.exp(c2 * x)

# Generar valores interpolados para graficar
x_values = np.linspace(3, 8, 100)
y_values = fitted_exponential(x_values)

# Graficar los puntos y la curva ajustada
plt.scatter(x_vals, y_vals, color="red", label="Datos")
plt.plot(x_values, y_values, color="blue", label="Curva ajustada")
plt.title("Ajuste Exponencial")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.grid(True)
plt.legend()
plt.show()

# Imprimir resultados
print(f'{round(c1, 3)} * e^({round(c2, 3)} * x)')
print(f'Error: {calculate_squared_error(y_vals, fitted_exponential, x_vals)}')
```

```python
### 1e construya el polinomio por minimos cuadrados de la forma bx^a y calcule el error  
# Definir la función potencial
def power_function(x, coef, exponent):
    return coef * x**exponent

# Ajustar los datos al modelo
params, _ = curve_fit(power_function, x_vals, y_vals)
coef, exponent = params

# Crear la función ajustada
def fitted_power_function(x):
    return coef * x**exponent

# Generar puntos para la curva ajustada
x_fit = np.linspace(3, 8, 100)
y_fit = fitted_power_function(x_fit)

# Graficar los datos y el ajuste
plt.scatter(x_vals, y_vals, color="red", label="Datos")
plt.plot(x_fit, y_fit, color="blue", label="Curva ajustada")
plt.title("Ajuste Potencial")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.legend()
plt.grid(True)
plt.show()

# Mostrar resultados
print(f'{round(coef, 3)} * x^{round(exponent, 3)}')
print(f'Error: {calculate_squared_error(y_vals, fitted_power_function, x_vals)}')
```

```python
### 2a construya el polinomio por minimos cuadrados del grado 1 y calcule el error 
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Función lineal (polinomio grado 1)
def linear_func(x, m, b):
    return m * x + b

# Error cuadrático medio (MSE)
def calculate_squared_error(y_real, y_pred):
    return np.mean((np.array(y_real) - np.array(y_pred))**2)

# Datos
x_points = np.array([0.2, 0.3, 0.6, 0.9, 1.1, 1.3, 1.4, 1.6])
y_points = np.array([0.050446, 0.098426, 0.33277, 0.72660,
                     1.0972, 1.5697, 1.8487, 2.5015])

# Ajuste por mínimos cuadrados
params, _ = curve_fit(linear_func, x_points, y_points)
slope, intercept = params

# Función ajustada
def fitted_func(x):
    return slope * x + intercept

# Puntos para la recta
x_interp = np.linspace(0, 2, 100)
y_interp = fitted_func(x_interp)

# Visualización
plt.scatter(x_points, y_points, label="Datos")
plt.plot(x_interp, y_interp, label="Ajuste lineal")
plt.title("Ajuste por mínimos cuadrados (grado 1)")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()
plt.show()

# Resultados
error = calculate_squared_error(y_points, fitted_func(x_points))

print(f"Polinomio ajustado: y = {slope:.3f}x + {intercept:.3f}")
print(f"Error cuadrático medio: {error:.6f}")
```

##### **3.	La siguiente tabla muestra los promedios de puntos del colegio de 20 especialistas en matemáticas y ciencias computacionales, junto con las calificaciones que recibieron estos estudiantes en la parte de matemáticas de la prueba ACT (Programa de Pruebas de Colegios Americanos) mientras estaban en secundaria. Grafique estos datos y encuentre la ecuación de la recta por mínimos cuadrados para estos datos.**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Modelo lineal
# y = m x + b
def linear_func(x, m, b):
    return m * x + b

# Error cuadrático medio (MSE)
def calculate_squared_error(y_real, y_pred):
    return np.mean((np.array(y_real) - np.array(y_pred))**2)

# Datos
ACT_scores = np.array([28, 25, 28, 27, 28, 33, 28, 29, 23, 27,
                       29, 28, 27, 29, 21, 28, 28, 26, 30, 24])

average_scores = np.array([3.84, 3.21, 3.23, 3.63, 3.75, 3.20, 3.41, 3.38,
                           3.53, 2.03, 3.75, 3.65, 3.87, 3.75, 1.66, 3.12,
                           2.96, 2.92, 3.10, 2.81])

# Ajuste por mínimos cuadrados
params, _ = curve_fit(linear_func, ACT_scores, average_scores)
m, b = params

# Función ajustada
def fitted_func(x):
    return m * x + b

# Puntos para la recta
x_interp = np.linspace(20, 34, 100)
y_interp = fitted_func(x_interp)

# Gráfica
plt.scatter(ACT_scores, average_scores, label="Datos")
plt.plot(x_interp, y_interp, label="Ajuste lineal")
plt.title("Ajuste lineal: ACT vs Promedio Académico")
plt.xlabel("Puntaje ACT")
plt.ylabel("Promedio Académico")
plt.grid(True)
plt.legend()
plt.show()

# Resultados
error = calculate_squared_error(average_scores, fitted_func(ACT_scores))

print(f"Modelo ajustado: y = {m:.3f}x + {b:.3f}")
print(f"Error cuadrático medio: {error:.6f}")
```

##### **4.	El siguiente conjunto de datos, presentando el Subcomité Antimonopolio del Senado, muestra las características comparativas de supervivencia durante un choque de automóviles de diferentes clases. Encuentre la recta por mínimos cuadrados que aproxima estos datos (la tabla muestra el porcentaje de vehículos que participaron en un accidente en los que la lesión más grave fue fatal o seria)**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Modelo lineal
# y = m x + b
def linear_func(x, m, b):
    return m * x + b

# Error cuadrático medio (MSE)
def calculate_squared_error(y_real, y_pred):
    return np.mean((np.array(y_real) - np.array(y_pred))**2)

# Datos
average_weights = np.array([4800, 3700, 3400, 2800, 1900])
accident_percentages = np.array([3.1, 4.0, 5.2, 6.4, 9.6])

# Ajuste por mínimos cuadrados
params, _ = curve_fit(linear_func, average_weights, accident_percentages)
m, b = params

# Función ajustada
def fitted_func(x):
    return m * x + b

# Puntos para la recta
x_interp = np.linspace(1800, 5000, 100)
y_interp = fitted_func(x_interp)

# Gráfica
plt.scatter(average_weights, accident_percentages, label="Datos")
plt.plot(x_interp, y_interp, label="Ajuste lineal")
plt.title("Ajuste lineal: Peso promedio vs Porcentaje de accidentes")
plt.xlabel("Peso promedio")
plt.ylabel("Porcentaje de accidentes")
plt.grid(True)
plt.legend()
plt.show()

# Resultados
error = calculate_squared_error(accident_percentages, fitted_func(average_weights))

print(f"Modelo ajustado: y = {m:.3f}x + {b:.3f}")
print(f"Error cuadrático medio: {error:.6f}")
```
