# Insertar Fotografia 


````python
import random

def bubble_sort(arr):
    n = len(arr)
    counter = 0  
    for i in range(n):
        swapped = False
        for j in range(0, n - i - 1):
            counter += 1  
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                swapped = True
        if not swapped:
            break
    return arr, counter

v2 = [-1, 0, 4, 5, 6, 7]
sorted_v2, counter_v2 = bubble_sort(v2)
print("v2_sorted:", sorted_v2)
print("Comparaciones en v2:", counter_v2)

v3 = [random.randint(-200, 145) for _ in range(100000)]
sorted_v3, counter_v3 = bubble_sort(v3)
print("v3_sorted:", sorted_v3[:10])  
print("Comparaciones en v3:", counter_v3)
````
