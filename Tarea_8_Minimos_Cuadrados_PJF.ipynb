#### 1A Construya el polinomio de minimos cuadrados de grado 1 y calcule el error.

```python


import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Datos
x_vals = [4.0, 4.2, 4.5, 4.7, 5.1, 5.5, 5.9, 6.3, 6.8, 7.1]
y_vals = [102.56, 130.11, 113.18, 142.05, 167.53, 195.14, 224.87, 256.73, 299.50, 326.72]

# definir  la función lineal
def linear_func(x, slope, intercept):
    return slope * x + intercept

# ajuste de parametros

params, _ = curve_fit(linear_func, x_vals, y_vals)
slope, intercept = params

# creacion de la funcion ajustada
def fitted_func(x):
    return slope * x + intercept

# calcular puntos de la línea ajustada
x_interp = np.linspace(3, 8, 100)
y_interp = fitted_func(x_interp)

# calculo del error cuadratico
def calculate_squared_error(observed, predicted_func, x_data):
    total_error = 0
    for i in range(len(observed)):
        total_error += (observed[i] - predicted_func(x_data[i])) ** 2
    return total_error

# Graficar los datos
plt.scatter(x_vals, y_vals, color='blue', label="Datos")
plt.plot(x_interp, y_interp, color='red', label="linea ajustada")
plt.title("ajuste-lineal")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.legend()
plt.show()  

# Resultados
print(f'{round(slope, 3)} x + {round(intercept, 3)}')
print(f'EL Error es de: {calculate_squared_error(y_vals, fitted_func, x_vals)}')
```

#### 1b.	Construya el polinomio de mínimos cuadrados de grado 2 y calcule el error.

```python


# definir la funcion cuadrática
def quadratic_eq(x, b2, b1, b0):
    return b2 * x**2 + b1 * x + b0

# ajuste del modelo cuadratico
coeffs, _ = curve_fit(quadratic_eq, x_vals, y_vals)
b2, b1, b0 = coeffs

# crear la función ajustada
def fitted_quadratic(x):
    return b2 * x**2 + b1 * x + b0

# generar datos interpolados
x_fit = np.linspace(3, 8, 100)
y_fit = fitted_quadratic(x_fit)


# graficar la curva con los datos 
plt.scatter(x_vals, y_vals, color='blue', label="Datos")
plt.plot(x_fit, y_fit, color='red', label="curva-ajustada")
plt.title("ajuste-cuadratico")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.grid(True)
plt.legend()
plt.show()

# Imprimir resultados
print(f'{round(b2, 3)} x^2 + {round(b1, 3)} x + {round(b0, 3)}')
print(f'El error es de : {calculate_squared_error(y_vals, fitted_quadratic, x_vals)}')
```

#### 1c.	Construya el polinomio de mínimos cuadrados de grado 3 y calcule el error

```python


# Definir la función cubica
def cubic_function(x, coef3, coef2, coef1, coef0):
    return coef3 * x**3 + coef2 * x**2 + coef1 * x + coef0

# ajustar el modelo cubico
coefficients, _ = curve_fit(cubic_function, x_vals, y_vals)
coef3, coef2, coef1, coef0 = coefficients

# definir la función ajustada
def adjusted_cubic(x):
    return coef3 * x**3 + coef2 * x**2 + coef1 * x + coef0

# generar los valores interpolados
x_interp = np.linspace(3, 8, 100)
y_interp = adjusted_cubic(x_interp)

# graficar los datos y la curva ajustada
plt.scatter(x_vals, y_vals, color="blue", label="Datos")
plt.plot(x_interp, y_interp, color="red", label="curva-ajustada")
plt.title("ajuste-cubico")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.grid(True)
plt.legend()
plt.show()

# Imprimir resultados
print(f'{round(coef3, 3)} x^3 + {round(coef2, 3)} x^2 + {round(coef1, 3)} x + {round(coef0, 3)}')
print(f'El error es de: {calculate_squared_error(y_vals, adjusted_cubic, x_vals)}')
```

##### 1d Construya el polinomio de mínimos cuadrados de la forma be^ax y calcule el error

```python


# definir la funcion exponencial
def exponential_func(x, c1, c2):
    return c1 * np.exp(c2 * x)

# ajustar el modelo exponencial
coefficients, _ = curve_fit(exponential_func, x_vals, y_vals)
c1, c2 = coefficients

# crear la función ajustada
def fitted_exponential(x):
    return c1 * np.exp(c2 * x)

# generar valores interpolados para graficar
x_values = np.linspace(3, 8, 100)
y_values = fitted_exponential(x_values)

# graficar los puntos y la curva ajustada
plt.scatter(x_vals, y_vals, color="red", label="Datos")
plt.plot(x_values, y_values, color="blue", label="curva-ajustada")
plt.title("ajuste-exponencial")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.grid(True)
plt.legend()
plt.show()

# imprimir resultados
print(f'{round(c1, 3)} * e^({round(c2, 3)} * x)')
print(f'El error es de: {calculate_squared_error(y_vals, fitted_exponential, x_vals)}')
```

#### 1e contruya el polinomio de minimos cuadrados de la formabx^a y calcule el error.

```python


# definir la funcion potencial
def power_function(x, coef, exponent):
    return coef * x**exponent

# ajustar los datos al modelo
params, _ = curve_fit(power_function, x_vals, y_vals)
coef, exponent = params

# crear la funcion ajustada
def fitted_power_function(x):
    return coef * x**exponent

# generar puntos para la curva ajustada
x_fit = np.linspace(3, 8, 100)
y_fit = fitted_power_function(x_fit)

# graficar los datos y el ajuste
plt.scatter(x_vals, y_vals, color="red", label="Datos")
plt.plot(x_fit, y_fit, color="blue", label="curva-ajustada")
plt.title("ajuste-potencial")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.legend()
plt.grid(True)
plt.show()

# imprimr resultados
print(f'{round(coef, 3)} * x^{round(exponent, 3)}')
print(f'El error es de: {calculate_squared_error(y_vals, fitted_power_function, x_vals)}')
```
